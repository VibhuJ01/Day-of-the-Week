```python
days = ["Sunday","Moday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
months = ["January","Febuary","March","April","May","June","July","August","September","October","November","December"]


#To check odd days in given century
def check_century(year):

    if(year <= 100):
        odd_days = 5
        
    elif((year-1)%400 < 100):
        odd_days = 0

    elif((year-1)%400 < 200):
        odd_days = 5

    elif((year-1)%400 < 300):
        odd_days = 3
    
    else:
        odd_days = 1

    return odd_days



#To check odd days in given leftover years
def check_year(year):

    year = (year-1)%100
    odd_days = 2*(year//4) + (year - year//4)

    return odd_days%7



#To check odd days in given months
def check_month(month,day,year):
    
    m = [3,0,3,2,3,2,3,3,2,3,2,3]
    sum = 0
    for i in range(month-1):
        sum += m[i]

    if(year%4 == 0 and month > 2):
        sum+=1
        
    sum += day%7

    return sum%7
    
    
def main():
    print("\n--------------------------------------------\n")
    print("This progam is to check day of the week on any given date in A.D.")

    ans = "y"

    while(ans.lower() == "y"):
        print("\n--------------------------------------------\n")

        #Taking Inputs
        month = int(input("Enter Month(1-12): "))
        day = int(input("Enter Day: "))
        year = int(input("Enter Year: "))

        #Finding out odd days
        odd_days = check_century(year)
        odd_days += check_year(year)
        odd_days += check_month(month,day,year)

        #Result
        print("\nDay of the week on",day,months[month-1],",",year,"was",days[odd_days%7])
        
        print("\n--------------------------------------------\n")
        ans = input("Do you want to try different values(y/n): ")

    print("\n--------------------------------------------\n")
    3print("Thank You")
    print("\n--------------------------------------------\n")

if(__name__ == "__main__"):
    main()
```
